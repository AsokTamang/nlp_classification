Reading the data

```python
df = pd.read_csv('/kaggle/input/competitions/titanic/train.csv')
df.head()
```

Data Cleaning

```python
df.isna().sum()
```

Extracting the highest frequently occuring data for each features

```python
df_mode = df.mode().iloc[0]
print('The highest frequently occuring data for each columns is: \n',df_mode)
```

Filling the null values of each feature with the mode value of each corresponding feature

```python
df = df.fillna(df_mode)
```
